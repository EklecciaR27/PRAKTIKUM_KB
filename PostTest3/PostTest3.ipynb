```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("MoviesOnStreamingPlatforms.csv")
```

```python
df.head(10)
```

## Grafik Batang - Seaborn

```python
df.info()
```

```python
df["Year"] = pd.to_datetime(df["Year"])
df.info()
```

```python
plt.figure(figsize=(20,20))
plt.title("Grafik Data Film")
plt.xlabel("Kategori Umur")
sns.countplot(data=df.head(1000), x="Age", palette='flare', hue='Year')
plt.show()
```

#### Dapat disimpulkan bahwa dari tahun ke tahun, Produksi Film kebanyakan memproduksi Film dengan kategori 18+ dan puncaknya terdapat pada tahun 2017. Kemudian Produksi Film dengan kategori SU (Semua Umur) Sangat minim.. Lalu Peringkat kedua film dengan ketegori usia terbanyak ada pada kaetgori 13+

## Grafik ScatterPlot


```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df.head(1000), x="Year", y="Age", hue="Age")
plt.show()
```

## Grafik PIE

```python
data_pie = df['Age'].value_counts().rename('Age').reset_index(name = 'id_count')
data_pie.head()
```

```python
plt.pie(
    data_pie["id_count"], 
    labels=data_pie["index"], 
    autopct="%.1f%%", 
    explode=(0.01 , 0.01 , 0.01 , 0.01 , 0.01 ),
    shadow=True 
) 
plt.show()
```

#### Kesimpulan dari Grafik Pie Diatas adalah, Kategori umur di dalam industri perfilman sendiri yaitu mayoritas film memiliki kategori usia 18+ dengan persentase 42.6%, kemudian diikuti oleh kategori 7+.
