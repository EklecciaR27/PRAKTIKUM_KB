### Steategi 1 : Subsitusi Nilai NULL (SKLEARN)


```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy = "mean")
df_null['ALP'] = imputer.fit_transform(df_null[['ALP']])
df_null['ALB'] = imputer.fit_transform(df_null[['ALB']])
df_null['ALT'] = imputer.fit_transform(df_null[['ALT']])
df_null['CHOL'] = imputer.fit_transform(df_null[['CHOL']])
df_null['PROT'] = imputer.fit_transform(df_null[['PROT']])
df_null['CREA'] = imputer.fit_transform(df_null[['CREA']])
df_null['GGT'] = imputer.fit_transform(df_null[['GGT']])
df_null.isna().sum()
```

### Strategi 2 : Subsitusi Nilai NULL (Pandas)

```python
df['ALP'] = df['ALP'].fillna(df['ALP']).mean()
df['ALB'] = df['ALB'].fillna(df['ALB']).mean()
df['ALT'] = df['ALT'].fillna(df['ALT']).mean()
df['CHOL'] = df['CHOL'].fillna(df['CHOL']).mean()
df['PROT'] = df['PROT'].fillna(df['PROT']).mean()
df['CREA'] = df['CREA'].fillna(df['CREA']).mean()
df['GGT'] = df['GGT'].fillna(df['GGT']).mean()
```

```python
df.isna().sum()
```

### Strategi 3 : DROPNA() 

```python
df.dropna(axis=0)
```
